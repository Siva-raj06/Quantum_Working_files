```python
#Creating a Quantum Circuit 
from qiskit import QuantumCircuit 
qc = QuantumCircuit(1) 
#Create a Quantum Circuit with 1 qubit and 0 classical bits.
#This initializes the qubit in the ground state |0> by default.
qc.data
```

```python
# ---------------------------------------------------------
# IMPORT: Statevector class from the quantum_info module
# ---------------------------------------------------------
from qiskit.quantum_info import Statevector

# ACTION: Create a statevector using a convenient text label ('0' or '1')
# PERFORMANCE: This is a high-level helper method. It automatically 
# generates the correct normalized computational basis vector [1, 0].
sv = Statevector.from_label('0')

# ACTION: Create a statevector using raw numerical input (Nested List)
# PERFORMANCE: This is a low-level definition. It manually defines the 
# amplitudes. Qiskit interprets [[1],[0]] as a column vector where 
# the amplitude of |0> is 1 and |1> is 0.
sv0 = Statevector([[1],[0]])

# ACTION: Standard Python print
# PERFORMANCE: Outputs the raw data structure (array) for debugging.
print(sv)
print(sv0)

# ---------------------------------------------------------
# VISUALIZATION: Prettifying abstract math
# ---------------------------------------------------------
from qiskit.visualization import array_to_latex

# ACTION: Render the vectors as LaTeX equations
# PERFORMANCE: array_to_latex turns the numerical array into a 
# mathematical matrix format. We use display() instead of print() 
# because it allows the Jupyter/OneNote engine to render the LaTeX 
# symbols properly instead of showing raw string text.
display(array_to_latex(sv))
display(array_to_latex(sv0))
```

```python
#now we apply the quantum circuit to this state by evolve method using the follwing
new_sv = sv.evolve(qc)
#you can look at the new statevector by writing
print(new_sv)
display(new_sv)
display(array_to_latex(new_sv))
#the statevector hasn't changed
```

```python
#can compute the projection of new_sv onto sv below by writing
# ---------------------------------------------------------
# IMPORT: state_fidelity class from the quantum_info module
# ---------------------------------------------------------
from qiskit.quantum_info import state_fidelity
state_fidelity(new_sv,sv)
#The projection of new_sv onto sv is 1, indicating that the two states are identical
```

```python
# VISUALIZATION: Mapping a 2D Vector to 3D Space 
# ---------------------------------------------------------
# NOTE: The correct function name in Qiskit is plot_bloch_multivector 
# or plot_state_bloch (depending on version). 
from qiskit.visualization import plot_bloch_multivector
display(plot_bloch_multivector(new_sv))
```

```python
#Now We are checking for State or ket1 vector
sv = Statevector.from_label('1')
new_sv = sv.evolve(qc)
display(array_to_latex(new_sv)) #display the vector 
display(state_fidelity(new_sv,sv))#checking the projection
display(plot_bloch_multivector(new_sv))#displaying the vector on bloch sphere
```

```python
#now creating a new QC->QuantumCircuit
QC = QuantumCircuit(1)
#now Creating a state vector of ket0 and ket1
QC.x(0)#x means pauli-x matrix
QC.draw('mpl')#check the circuit below we perform X operation to the quantum circuit
```

```python
sv0 = Statevector.from_label('0')
sv1 = Statevector.from_label('1')
print("ket0 vector before operation and on the bloch sphere")
display(array_to_latex(sv0))
display(plot_bloch_multivector(sv0))
print("ket1 vector before operation and on the bloch sphere")
display(array_to_latex(sv1))
display(plot_bloch_multivector(sv1))
#now we are going with x gate apply to vector of ket0 and ket1 individually
new_sv0 = sv0.evolve(QC)
new_sv1 = sv1.evolve(QC)
#displaying the new vector
print("ket0 vector after operation and on the bloch sphere")
display(array_to_latex(new_sv0))
display(plot_bloch_multivector(new_sv0))
print("ket1 vector after operation and on the bloch sphere")
display(array_to_latex(new_sv1))
display(plot_bloch_multivector(new_sv1))
```

```python
'''This unitary operation R is an interesting example. By applying this operation twice, which is equivalent to squaring its matrix representation, we obtain a NOT operation'''
circuit = QuantumCircuit(1)
circuit.h(0)
circuit.s(0)
circuit.h(0)
circuit.h(0)
circuit.s(0)
circuit.h(0)
circuit.draw('mpl')
```

```python
sv0 = Statevector.from_label('0')
sv1 = Statevector.from_label('1')
print("ket0 vector before operation and on the bloch sphere")
display(array_to_latex(sv0))
display(plot_bloch_multivector(sv0))
print("ket1 vector before operation and on the bloch sphere")
display(array_to_latex(sv1))
display(plot_bloch_multivector(sv1))
#now we are going with HSHHSH gate apply to vector of ket0 and ket1 individually
new_sv0 = sv0.evolve(circuit)
new_sv1 = sv1.evolve(circuit)
#displaying the new vector
print("ket0 vector after operation and on the bloch sphere")
display(array_to_latex(new_sv0))
display(plot_bloch_multivector(new_sv0))
print("ket1 vector after operation and on the bloch sphere")
display(array_to_latex(new_sv1))
display(plot_bloch_multivector(new_sv1))
```

```python
from qiskit.quantum_info import Operator
from numpy import sqrt

Y = Operator([[0, -1.0j], [1.0j, 0]])
H = Operator([[1 / sqrt(2), 1 / sqrt(2)], [1 / sqrt(2), -1 / sqrt(2)]])
S = Operator([[1, 0], [0, 1.0j]])
T = Operator([[1, 0], [0, (1 + 1.0j) / sqrt(2)]])

display(T.draw("latex"))
```

```python
v = Statevector([1, 0])

v = v.evolve(H)
v = v.evolve(T)
v = v.evolve(H)
v = v.evolve(S)
v = v.evolve(Y)

display(v.draw("latex"))
```

```python
from qiskit_aer import AerSimulator

# 1. Create a circuit with a measurement
qc_measure = QuantumCircuit(1)
qc_measure.h(0)        # Put into superposition
qc_measure.measure_all() # Add measurement gates

# 2. Use the Simulator to run the circuit
backend = AerSimulator()
result = backend.run(qc_measure, shots=1024).result()

# 3. Get the counts (how many times we saw '0' vs '1')
counts = result.get_counts()
print(f"Measurement results: {counts}")

from qiskit.visualization import plot_histogram
display(plot_histogram(counts))
```
