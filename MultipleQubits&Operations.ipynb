```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector,Operator
from qiskit.visualization import array_to_latex,plot_bloch_multivector,plot_state_qsphere
from numpy import sqrt
```

```python
#we are creating a state vectors of 0 and 1
sv1= Statevector.from_label('0')
sv2 = Statevector.from_label('1')
display(array_to_latex(sv1))
display(array_to_latex(sv2))
#now we are going to do tensor product of |0> and |1> and vice-versa
psi = sv1.tensor(sv2)
psi0 =sv2.tensor(sv1)
#state formed after doing tensor with |0>*|1>
display("state of |0>*|1>:")
display(psi.draw("latex"))
display(array_to_latex(psi))
display(plot_bloch_multivector(psi))
display(plot_state_qsphere(psi))
display("state of |1>*|0>:")
display(psi0.draw("latex"))
display(array_to_latex(psi0))
display(plot_bloch_multivector(psi0))
display(plot_state_qsphere(psi0))
```

```python
plus = Statevector.from_label("r")#instead of + we use r or we can use + too
minus_i = Statevector.from_label("l")#instead of - we use l or we dont use - because
minus = Statevector.from_label("-")
display(minus.draw("latex"))#there is no conjujate in the real -
display(minus_i.draw("latex"))#so we are using l
phi = plus^minus_i #instead of tensor written we use ^ for tensor two vectors
display(phi.draw("latex"))
display(plot_state_qsphere(phi))
display(plot_bloch_multivector(phi))
```

```python
# 1. Initialize a 2-qubit Quantum Circuit
QC = QuantumCircuit(2)

# 2. Create entanglement (Bell State setup)
# Apply Hadamard gate to qubit 0 to create superposition: (|0> + |1>) / sqrt(2)
QC.h(0)
# Apply CNOT gate with qubit 0 as control and qubit 1 as target
# This links the qubits, creating the Bell State: (|00> + |11>) / sqrt(2)
QC.cx(0,1)

# Draw the circuit to verify the gate sequence
QC.draw('mpl')

# 3. Define the four basis states in the computational basis
sv0 = Statevector.from_label('00')
sv1 = Statevector.from_label('01')
sv2 = Statevector.from_label('10')
sv3 = Statevector.from_label('11')

# 4. Evolve each basis state through the circuit
# This shows how the circuit transforms different possible inputs
psi0 = sv0.evolve(QC) # Results in (|00> + |11>) / sqrt(2)
psi1 = sv1.evolve(QC) # Results in (|01> + |10>) / sqrt(2)
psi2 = sv2.evolve(QC) # Results in (|00> - |11>) / sqrt(2)
psi3 = sv3.evolve(QC) # Results in (|01> - |10>) / sqrt(2)

# 5. Visualization
# Displaying states in Bra-Ket notation (standard quantum notation)
# Displaying states as column vectors (linear algebra representation)
# Note: Ensure you pass the variable (psi0), not a string ("psi0")
display(psi0.draw("latex"))
display(array_to_latex(psi0, prefix="\\psi_0 = "))
display(psi1.draw("latex"))
display(array_to_latex(psi1, prefix="\\psi_1 = "))
display(psi2.draw("latex"))
display(array_to_latex(psi2, prefix="\\psi_2 = "))
display(psi3.draw("latex"))
display(array_to_latex(psi3, prefix="\\psi_3 = "))
```
